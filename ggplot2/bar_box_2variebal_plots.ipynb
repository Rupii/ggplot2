## BoxPlot

```r
library(ggplot2)
df <- mtcars
head(df)
```

```r
pl <- ggplot(df, aes(x = factor(cyl), y = mpg))
print(pl + geom_boxplot() + coord_flip())
# box plots on x axis should have categorical values
# co ordinate is used to chage the coordinate of the plot
# to know what it exactly means remove coord_fil() and see yourself|
```

```r
pl <- ggplot(df, aes(x = factor(cyl), y = mpg))
print(pl + geom_boxplot(aes(fill = factor(cyl))) + theme_dark())
```
