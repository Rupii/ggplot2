```r
library(ggplot2)
head(mpg)
```

```r
pl <- ggplot(mpg, aes(displ , hwy)) + geom_point()
print(pl)
```

## Coordinates

```r
pl + coord_cartesian(xlim = c(1, 5))
# the coordinates take vetors from lower limit to heigher
```

```r
pl + coord_fixed(ratio = 1/3)
```

## facet

```r
print(pl + facet_grid(. ~cyl))
# . ~ cyl represent y ~ x
# devided x and y in into facet
# above dot is dummy if you want to facet by  one variable
```
