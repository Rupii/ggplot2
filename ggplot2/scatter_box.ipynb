# R in Action 
### ggplot2 plots

Scatter plot with regression line

```r
require(ggplot2)
# loading ggplot
head(mtcars)
# loading mtcars datasets
```

```r
pl <- ggplot(mtcars, aes(wt, mpg))
print(pl + geom_point(col = '#00f0ff',pch =16))
```

```r
pl + geom_point(col = '#00f0ff',pch =17) -> q
print(q + geom_smooth(col = 'red', method = lm, linetype = 2 ))
```

### Box Plot

```r
#install.packages("cars")
#library(cars)
data(Salaries, package="car")
head(Salaries)
```

```r
pl <- ggplot(Salaries, aes(rank, salary)) 
print(pl + geom_boxplot() )
```

```r
q <- pl +geom_boxplot(fill="cornflowerblue", color="black", notch=TRUE) + geom_point(position="jitter", color="blue", alpha=.5) + geom_rug(side="l", color="black")
print(q)
```
