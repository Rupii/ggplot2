```r
ggplot(movies, aes(x = rating))  -> p
print(p + geom_histogram(binwidth = 0.1))
```

```r
q <- ggplot(movies, aes(x = rating)) + geom_histogram(binwidth = 0.1, col = "red", fill = 'blue', alpha = 0.5)
# only key here is the alpha which stands for the transpiracy of the histogram
print(q)
```

```r
plt <- q + xlab("Movie Rantings") + ylab("give y here") 
print(plt + ggtitle("Movies Hist"))
#the lable goes under the facet layer
```

## Scatter Plots

```r
library(ggplot2)
df <- mtcars
head(df)
```

```r
# we can assign the varibles either way that one of advantage of R
ggplot(df, aes(x = wt, y = mpg)) -> pl
print(pl + geom_point())
```

```r
print(pl + geom_point(size = 6, alpha = 0.5))
```

```r
# we can control size as a varible using aes() in geom layer i.e.,
pl + geom_point(aes(size = hp, col = hp)) -> plt
print(plt + scale_color_gradient(low = 'blue', high = 'red'))

# you can also use the standard # rgb colors in ggplot
```
